Yêu cầu 
Xây dựng lớp Shape với phương thức area(). 
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area(). 
Lấy π = 3.14. 
 
Input 
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE) 
Dòng 2: tham số tương ứng. 
 
Output 
In ra diện tích với 2 chữ số thập phân.

```python
import math

class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    
    def area(self):
        return 3.14 * (self.r ** 2)

shape_type = input().strip().upper()
params = list(map(float, input().split()))

if shape_type == "RECTANGLE":
    obj = Rectangle(params[0], params[1])
    print(f"{obj.area():.2f}")
elif shape_type == "CIRCLE":
    obj = Circle(params[0])
    print(f"{obj.area():.2f}")
```

số nguyên dương x. 
 
Input 
Một số nguyên dương n. 
 
Output 
In ra dạng nhị phân của n.

```python
class Printer:
    def print_value(self, x):
        print(x)

class BinaryPrinter(Printer):
    def print_value(self, x):
  
        
        print(bin(int(x))[2:])

# Input: Một số nguyên dương n
try:
    n = input().strip()
    p = BinaryPrinter()
    p.print_value(n)
except EOFError:
    pass
```

Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận: 
[a b] 
[c d] 
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc. 
 
Input 
Một dòng chứa 4 số nguyên a b c d. 
 
Output 
In ra trace và determinant trên cùng một dòng

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a, self.b, self.c, self.d = a, b, c, d
    
    def trace(self):
        return self.a + self.d
    
    def determinant(self):
        return (self.a * self.d) - (self.b * self.c)

try:
    nums = list(map(int, input().split()))
    m = Matrix2x2(*nums)
    print(f"{m.trace()} {m.determinant()}")
except:
    pass
```

In ra trace và determinant trên cùng một dòng.

Yêu cầu 
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), 
pop(), top(). 
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ. 
 
Input 
Dòng 1: n 
n dòng tiếp theo là một trong ba lệnh: 
PUSH x 
POP 
TOP 
 
Output 
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, x):
        self.items.append(x)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
    
    def top(self):
        if not self.is_empty():
            return self.items[-1]
            
    def is_empty(self):
        return len(self.items) == 0

try:
    n = int(input())
    s = Stack()
    for _ in range(n):
        line = input().split()
        cmd = line[0]
        
        if cmd == "PUSH":
            s.push(line[1])
        elif cmd == "POP":
            s.pop()
        elif cmd == "TOP":
            print(s.top())
except:
    pass
```
