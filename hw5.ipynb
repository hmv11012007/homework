Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def is_amstrong(x: int) ->bool:
    if x<0:
        return False
    s= str(x)
    n=len(s)
    tong= sum(int(digit)**n for digit in s)
    return tong==x
num=int(input( ))
if is_amstrong(num):
    print(f"{num} là một số amstrong")
else:
    print(f"{num} ko là một số amstrong")
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
import math
def is_perfect_square(x:int) ->bool:
    if x<0:
        return False
    n=math.isqrt(x)
    return n*n==x
num=int(input())
if is_perfect_square(num):
    print(f"{num} là số chính phương")
else:
    print(f"{num} ko là số chính phương")
```

VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_palindrome(x:int) -> bool:
    if x<0 or (x%10==0 and x !=0):
        return False
    original_num=x
    revered_num=0
    while x>0:
        digit=x%10
        revered_num=revered_num*10 + digit
        x//=10
        return revered_num == original_num
    
    
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def is_one_product_of_others(a: int, b: int, c: int) -> bool:
  
    case1 = (a == b * c)
    case2 = (b == a * c)
    case3 = (c == a * b)

    return case1 or case2 or case3
print(is_one_product_of_others(6, 2, 3))   
print(is_one_product_of_others(2, 10, 5))  
print(is_one_product_of_others(3, 4, 7))   
print(is_one_product_of_others(0, 0, 5))  
```

Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối. 

```python
def sum_halves(s: str) -> int:
   
    first_half = int(s[:3])
    
   
    second_half = int(s[3:])
    

    return first_half + second_half

# Kiểm tra thử
print(sum_halves("123456"))  # 123 + 456 = 579
print(sum_halves("001002"))  # 1 + 2 = 3
print(sum_halves("500500"))  # 500 + 500 = 1000
```

Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            return -1 if c == 0 else 0 
        return 1

   
    delta = b**2 - 4*a*c
    
  
    if delta > 0:
        return 2    
    elif delta == 0:
        return 1    
    else:
        return 0    
```

 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    val = (x2 - x1) * (y3 - y1) - (y2 - y1) * (x3 - x1)
    
    return val == 0


print(are_collinear(1, 1, 2, 2, 3, 3)) 
print(are_collinear(1, 1, 2, 2, 3, 4)) 
print(are_collinear(0, 0, 0, 5, 0, 10)) #
```

```python
def xor_operation(a: int, b: int) -> int:
    """
    Trả về kết quả của phép toán a XOR b.
    """
    return a ^ b

print(xor_operation(5, 3))   
print(xor_operation(10, 10)) 
print(xor_operation(0, 1))   
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]. 

```python
def get_max_min(a, b, c):
    gia_tri_max = max(a, b, c)
    gia_tri_min = min(a, b, c)
    return [gia_tri_max, gia_tri_min]
print(get_max_min(10, 5, 20))   
print(get_max_min(-1, -5, 0))  
print(get_max_min(7, 7, 7))     
```

Viết hàm tính tổng bình phương các số từ 1 đến n. 

```python
def sum_squares_loop(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
        total += i * i
    return total
```
