Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
from collections import Counter

def solve_w7a16(arr):
    counts = Counter(arr)
    removals = 0
    for x, count in counts.items():
        if count >= x:
            removals += (count - x)
        else:
            removals += count
    return removals

```

Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
def boyer_moore(arr):
    candidate = None
    count = 0
    for x in arr:
        if count == 0:
            candidate = x
            count = 1
        elif x == candidate:
            count += 1
        else:
            count -= 1
    
    # Kiểm tra lại xem có thực sự > n/2 không
    if arr.count(candidate) > len(arr) / 2:
        print(candidate)

```

Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng 
dần, lẻ giảm dần; ;

```python
def custom_sort(arr):
    evens = sorted([x for x in arr if x % 2 == 0])
    odds = sorted([x for x in arr if x % 2 != 0], reverse=True)
    return evens + odds

```

Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào

```python
def merge_sorted(a, b):
    # Đảm bảo cả 2 đều tăng dần để dễ xử lý
    if a[0] > a[-1]: a.reverse()
    if b[0] > b[-1]: b.reverse()
    
    res = []
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            res.append(a[i]); i += 1
        else:
            res.append(b[j]); j += 1
    return res + a[i:] + b[j:]
```

Cho dãy N số nguyên dương, tính tổng các số có k chữ số. Dòng đầu ghi

Cho N là ngày, T là tháng năm 2020, Tính xem đó là ngày thứ bao nhiêu 
của năm đó nếu đó là ngảy hợp lệ 

```python
import datetime

def day_of_year(d, m):
    try:
        date_obj = datetime.date(2020, m, d)
        return date_obj.timetuple().tm_yday
    except ValueError:
        return "Ngày không hợp lệ"

```

Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. 
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất 

```python
def most_frequent_name(names):
    counts = Counter(names)
    # Sắp xếp theo số lần xuất hiện giảm dần, sau đó theo bảng chữ cái tăng dần
    sorted_names = sorted(counts.items(), key=lambda x: (-x[1], x[0]))
    return sorted_names[0][0]
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài 
>=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí 
tự cái

Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên 

```python
def remove_adjacent_duplicates(s):
    words = s.split()
    if not words: return ""
    res = [words[0]]
    for i in range(1, len(words)):
        if words[i] != words[i-1]:
            res.append(words[i])
    return " ".join(res)
```

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số 

```python
def max_k_digits(s, k):
    stack = []
    to_remove = len(s) - k
    for digit in s:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    return "".join(stack[:k])

```

Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các 
số đó

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ 
hay không

```python
def is_valid_brackets(s):
    mapping = {')': '(', '}': '{', ']': '['}
    stack = []
    for char in s:
        if char in mapping.values():
            stack.append(char)
        elif char in mapping:
            if not stack or stack.pop() != mapping[char]:
                return False
    return len(stack) == 0
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau 

```python
def longest_cont_inc(arr):
    if not arr: return []
    max_sub = cur_sub = [arr[0]]
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            cur_sub.append(arr[i])
        else:
            cur_sub = [arr[i]]
        if len(cur_sub) > len(max_sub):
            max_sub = cur_sub
    return max_sub
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím

```python
def longest_increasing_subsequence(arr):
    if not arr: return []
    n = len(arr)
    f = [1] * n
    trace = [-1] * n
    for i in range(n):
        for j in range(i):
            if arr[j] < arr[i] and f[i] < f[j] + 1:
                f[i] = f[j] + 1
                trace[i] = j
    
    # Tìm vị trí kết thúc của LIS dài nhất
    max_idx = f.index(max(f))
    res = []
    while max_idx != -1:
        res.append(arr[max_idx])
        max_idx = trace[max_idx]
    return res[::-1]
```
