Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n=int(input())
a=n%10

b=n%100-a

c=n%1000-b-a

d=a*100+b+c/100
print(f"{d:.0f}")
```

Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm 
tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in 
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", 
"Tam giác cân", hay "Tam giác thường".

```python
import math
try:
    a = float(input("Nhập cạnh a: "))
    b = float(input("Nhập cạnh b: "))
    c = float(input("Nhập cạnh c: "))
    if (a + b > c) and (a + c > b) and (b + c > a):
        print("Đây là ba cạnh của một tam giác.")
        loai_tg = ""
        if a == b == c:
            loai_tg = "Đều"
        elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
            if a == b or b == c or a == c:
                loai_tg = "Vuông Cân"
            else:
                loai_tg = "Vuông"
            loai_tg = "Cân"
        else:
            loai_tg = "Thường"     
        print(f"Loại tam giác: {loai_tg}")
        
        
    else:
        print("Khong la ba canh cua mot tam giac")

except ValueError:
    print("Vui lòng nhập số hợp lệ.")
```
