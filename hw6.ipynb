Nhậ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
m,n=map(int,input().split( ))
matran=[]
for i in range(m):
    row=[]
    for j in range(n):
        val = int(input(f"Nhập phần tử hàng {i+1}, cột {j+1}: "))
        row.append(val)
    matran.append(row)
print("\nMa tran vua nhap la")
for row in matran:
    for item in row:
        # In phần tử với độ rộng 4 ký tự, căn phải
        print(f"{item:>4}", end="")
    # Xuống dòng sau khi in xong một hàng
    print()
```

Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
# Nhập kích thước ma trận vuông n x n
n = int(input("Nhập kích thước n của ma trận vuông: "))

# Khởi tạo ma trận
matrix = []

# Nhập dữ liệu cho ma trận
for i in range(n):
    row = []
    for j in range(n):
        val = int(input(f"Nhập phần tử [{i}][{j}]: "))
        row.append(val)
    matrix.append(row)

# In đường chéo chính
print("\nCác phần tử trên đường chéo chính là:")
for i in range(n):
    print(matrix[i][i], end=" ")

# In đường chéo phụ
print("\n\nCác phần tử trên đường chéo phụ là:")
for i in range(n):
    # Cột của đường chéo phụ tính bằng: n - 1 - i
    print(matrix[i][n - 1 - i], end=" ")
print()
```

Nhậ ̣ p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột 
k (hàng k) 

```python
# Nhập n, m, k trên cùng một dòng
data = input("Nhập n, m, k: ").split()
n = int(data[0])
m = int(data[1])
k = int(data[2])

# Khởi tạo ma trận
matrix = []

# Nhập dữ liệu cho ma trận
for i in range(n):
    row = list(map(int, input(f"Nhập hàng {i}: ").split()))
    matrix.append(row)

# 1. Tính tổng các phần tử trên hàng k
# Lưu ý: Trong lập trình, hàng đầu tiên thường bắt đầu từ chỉ số 0
if 0 <= k < n:
    tong_hang_k = sum(matrix[k])
    print(f"Tổng các phần tử trên hàng {k}: {tong_hang_k}")
else:
    print(f"Chỉ số hàng {k} không hợp lệ.")

# 2. Tính tổng các phần tử trên cột k
if 0 <= k < m:
    tong_cot_k = 0
    for i in range(n):
        tong_cot_k += matrix[i][k]
    print(f"Tổng các phần tử trên cột {k}: {tong_cot_k}")
else:
    print(f"Chỉ số cột {k} không hợp lệ.")
```

Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list 

```python
def tim_vi_tri_lon_nhat(lst):
    if not lst:
        return None
    
    
    gia_tri_max = max(lst)
    
    
    vi_tri = lst.index(gia_tri_max)
    
    return vi_tri


du_lieu = [10, 25, 7, 40, 15, 40]
ket_qua = tim_vi_tri_lon_nhat(du_lieu)

print(f"Danh sách: {du_lieu}")
print(f"Vị trí có giá trị lớn nhất là: {ket_qua}")
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về 

```python
def tim_vi_tri_k(lst, k):
    # Duyệt qua từng vị trí (index) trong danh sách
    for i in range(len(lst)):
        # Nếu giá trị tại vị trí i bằng k
        if lst[i] == k:
            return i  # Trả về vị trí và thoát hàm ngay lập tức
            
    # Nếu vòng lặp kết thúc mà không tìm thấy k
    return -1

# Ví dụ sử dụng:
ds_so = [10, 20, 30, 40, 50]
k_can_tim = 30

ket_qua = tim_vi_tri_k(ds_so, k_can_tim)

if ket_qua != -1:
    print(f"Giá trị {k_can_tim} xuất hiện đầu tiên tại vị trí: {ket_qua}")
else:
    print(f"Không tìm thấy giá trị {k_can_tim} trong danh sách.")
```

Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

 VIết hàm xóa đi các số <x có trong mảng 

```python
def xoa_phan_tu_nho_hon_x(lst, x):
    # Tạo một danh sách mới chỉ chứa các số >= x
    ket_qua = []
    
    for item in lst:
        if item >= x:
            ket_qua.append(item)
            
    return ket_qua

# Ví dụ sử dụng:
ds_so = [10, 5, 8, 20, 3, 15]
x = 10

ds_moi = xoa_phan_tu_nho_hon_x(ds_so, x)

print(f"Mảng ban đầu: {ds_so}")
print(f"Mảng sau khi xóa các số < {x}: {ds_moi}")
```

Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_khong_ve_dau(lst):
    # Lọc ra các số 0
    so_khong = [x for x in lst if x == 0]
    # Lọc ra các số khác 0 (vẫn giữ đúng thứ tự)
    so_khac = [x for x in lst if x != 0]
    
    # Ghép: số 0 đứng trước + số khác đứng sau
    return so_khong + so_khac

# Chạy thử
mang = [0, 2, 3, 2, 4, 0, 0, 9]
print("Đưa 0 về đầu:", day_khong_ve_dau(mang))
# Kết quả: [0, 0, 0, 2, 3, 2, 4, 9]
```

Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong 
mảng

```python
def tim_so_con_thieu(lst):
    cac_so_thieu = []
    
    # Duyệt qua các số từ 0 đến 9
    for i in range(10):
        # Nếu số i không xuất hiện trong danh sách
        if i not in lst:
            cac_so_thieu.append(i)
            
    return cac_so_thieu

# Ví dụ sử dụng:
mang_nhap = [1, 2, 4, 6, 7, 1, 2]
ket_qua = tim_so_con_thieu(mang_nhap)

print(f"Các số từ 0-9 không có trong mảng là: {ket_qua}")
# Kết quả: [0, 3, 5, 8, 9]
```

Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 
phần tử đầu và cuối 

```python
def tinh_tong_cuc_dai_dia_phuong(lst):
    tong = 0
    n = len(lst)
    
    # Duyệt từ vị trí thứ 1 đến n-2 (bỏ qua đầu 0 và cuối n-1)
    for i in range(1, n - 1):
        # Kiểm tra điều kiện cực đại địa phương
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
            
    return tong

# Ví dụ sử dụng:
# Các cực đại địa phương là: 8 (vị trí 2) và 10 (vị trí 4)
ds_so = [1, 5, 8, 3, 10, 2, 6] 
ket_qua = tinh_tong_cuc_dai_dia_phuong(ds_so)

print(f"Danh sách: {ds_so}")
print(f"Tổng các cực đại địa phương là: {ket_qua}")
# Giải thích: 8 + 10 = 18
```
