Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n=int(input())
if n>0 and (n & (n-1))==0:
    print(f"{n} có dạng 2^k")
else:
    print(f"{n} kho có dạng 2^k")
```

 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = input("Nhập vào số nguyên dương n: ")
uoc_chan = 0
uoc_le = 0
for chu_so in n:
    if int(chu_so) % 2 == 0:
        uoc_chan += 1
    else:
        uoc_le += 1

print(f"Số {n} có {uoc_chan} chữ số chẵn và {uoc_le} chữ số lẻ.")
```

Nhập n 
Cho S(k) = 1 + 2 + 3 + … + k 
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n=int(input())
k=0
s_k=0
while s_k + (k+1)<n:
    k +=1
    s_k +=k
print(f"Giá trị k lớn nhất là: {k}")
print(f"Tổng S({k}) = {s_k} (nhỏ hơn {n})")
```

Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A=float(input())
S=0
n=0
while S<A:
    n +=1
    S +=(1/n)
print(f"Số n nhỏ nhất tìm được là: {n}")
print(f"Tổng tương ứng S({n}) = {S:.5f} (> {A})")
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
so_max = None
so_min = None

while True:
    n = int(input("Nhập số: "))
    if n == -1:
        break
    
    if so_max is None or n > so_max:
        so_max = n
    if so_min is None or n < so_min:
        so_min = n

if so_max is not None:
    print(f"Max: {so_max}, Min: {so_min}")
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 

```python
A=int(input())
f1,f2=1,1
while True:
    f_tieptheo= f1+f2
    if f_tieptheo > A:
        break
    f1=f2
    f2=f_tieptheo
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
s = input("Nhập vào một chuỗi: ")
chu_hoa = 0
chu_thuong = 0
chu_so = 0
ky_tu_khac = 0
for char in s:
    if char.isupper():     
        chu_hoa += 1
    elif char.islower():   
        chu_thuong += 1
    elif char.isdigit():   
        chu_so += 1
    else:                    
        ky_tu_khac += 1
print(f"Số ký tự in hoa:    {chu_hoa}")
print(f"Số ký tự in thường: {chu_thuong}")
print(f"Số ký tự số:        {chu_so}")
print(f"Số ký tự khác:      {ky_tu_khac}")
```
