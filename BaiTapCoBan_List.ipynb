### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def kiem_tra_day_so(lst):
    ham_tang= True
    ham_giam= True

    
    for i in range(len(lst)-1):
        if lst[i] >= lst[i+1]:
            ham_tang= False
        if lst[i] <= lst[i+1]:
            ham_giam=False


    if ham_tang:
        return "Tang"
    elif ham_giam:
        return "giam"
    else:
        return "Vo danh"
    
   
   
lst=list(map(int,input().split( )))
print(kiem_tra_day_so(lst))

```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
def kiem_tra_ziczac(lst):
    n = len(lst)
    if n < 3:
        return True
    
    for i in range(1, n - 1):
        if (lst[i]-lst[i-1])*(lst[i+1]-lst[i]) <=0:
            return "la ham ziczac"
    return "khong laf ham ziczac"
             

lst=list(map(int,input().split( )))
print(kiem_tra_ziczac(lst))
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
def vi_tri_cua_k(lst,k):
    vi_tri=[i for i,x in enumerate(lst) if x==k]
    return vi_tri if vi_tri else "-1"
lst=[1,3,4,4,3,3,4,3,6,1]

k1=3
print(f"k={k1} : {vi_tri_cua_k(lst,k1)}")

k2=10
print(f"k={k2} : {vi_tri_cua_k(lst,k2)}")
```
