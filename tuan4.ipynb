Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000)

VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không?

```python
while True:
    try:
        n = int(input("Nhập một số nguyên dương: "))
        if n > 0:
            break
        else:
            print("Vui lòng nhập số lớn hơn 0!")
    except ValueError:
        print("Đó không phải là số nguyên. Thử lại!")

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
if is_prime(n):
    print(f"Số {n} là số nguyên tố.")
else:
    print(f"Số {n} không phải là số nguyên tố.")
```


Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) 
của số đó. Không dùng kiểu dữ liệu string. 

```python

n = int(input("Nhập số nguyên n: "))
bandau = n
n = abs(n)
count = 0
if n == 0:
    count = 1
else:
    while n > 0:
        n //= 10  
        count += 1
print(f"Số chữ số của {bandau} là: {count}")
```

Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

nhập vào số nguyên đến khi nhập số nguyên dương thì thôi kiểu tra đó có phải số nguyên tố không

```python
n=-10
while (n<=0):
    n=int(input())

snt=0 # biến đếm số ước
for i in range(1,n+1):
    if (n%i==0):
        snt +=1
if snt==2:
    print("Đây là số nguyên tố")
else :
    print("Đây là hợp số")
   
```

nhập vào a,b cho đến khi được số nguyên dương a,b và a<b
tính tổng số nguyên tố từ a đến b 
tính tổng các số nguyên tố trong đoạn [a,b]

```python
a=-10
b=-9
tongsnt=0
a=max(2,a)

while a<0 or b<0 or a<b:
    a,b=map(int,input().split( ))
for n in range(a, b+1):
    for i in range(2,int(n**0.5)+1):
        if n % i==0:
            break
        else:
            tongsnt = tongsnt + n
print(f"{tongsnt}")

        
```
